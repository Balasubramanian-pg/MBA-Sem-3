## Binning in Data Analysis

In data analysis, binning is a technique used to group data into distinct categories or "bins." This is done to simplify complex data, make it more manageable, and extract insights or patterns from it. Binning is particularly useful when dealing with continuous numerical data, where creating categories can provide a more structured view of the information.

### Why We Use Binning:

1. **Simplification**: Binning reduces the complexity of continuous data by grouping values into categories, making it easier to understand and interpret.

2. **Pattern Recognition**: Binning can help reveal underlying patterns or trends that might not be evident when dealing with individual data points.

3. **Analysis**: Grouping data into bins is often a prerequisite for various analysis techniques, such as creating histograms, bar charts, or pivot tables, which require discrete categories.

In the provided code, we are using binning to categorize 'Employment_Tenure(Months)' into specific bins, each representing a range of months. This allows us to analyze and understand the distribution of 'Credit_default' cases in different employment tenure groups. Binning is an essential step in the process of summarizing and exploring data effectively.


````python
# Creating Bins for 'Employment_Tenure(Months)' and a Pivot Table Analysis:

# We're creating bins to categorize 'Employment_Tenure(Months)' into different groups for analysis.

# Define the bin edges:
bins = [0, 20, 40, 60, float('inf')]

# Define the bin labels for each category:
labels = ['0-20', '20-40', '40-60', '60+']

# Bin the data and create a new column 'Employment_Tenure_Bin':
data['Employment_Tenure_Bin'] = pd.cut(data['Employment_Tenure(Months)'], bins=bins, labels=labels)

# Now, we create a pivot table to examine the count of 'Credit_default' cases based on 'Employment_Tenure_Bin' and 'Stable_Income':

# Create the pivot table using the new 'Employment_Tenure_Bin' column:
pivot_table_4 = data.pivot_table(index='Employment_Tenure_Bin', columns='Stable_Income', values='Credit_default', aggfunc='count')

# Print a descriptive title for the pivot table:
print("Pivot Table 4: Credit Default by Employment Tenure and Stable Income")

# Display the pivot table, which shows the count of 'Credit_default' cases for each combination of 'Employment_Tenure_Bin' and 'Stable_Income'.
print(pivot_table_4)
````

## Hypothesis 1: Young Borrowers (22-32) and Loan Defaults

### Technique Used: Histogram Analysis

- For Hypothesis 1, we employed a histogram analysis to explore the relationship between borrower age and loan defaults.
- This technique was preferred because it provides a clear visual representation of the distribution of credit defaults within different age groups. Histograms are effective in identifying patterns and differences in data.

### Data Representation:
- The data was represented using histograms that separate individuals who defaulted on loans ('Yes') from those who did not ('No').
- The x-axis represents different age groups, and the y-axis shows the count of individuals in each group.

### Hypothesis Test:
- Based on the histogram analysis, we observed that young borrowers in the age group of 22-32 had a higher count of loan defaults ('Yes') compared to older borrowers. Therefore, the hypothesis was tested as **True**.


````python
# Visualizing Credit Default by Age Group:

# Create a figure with a size of 12x6 inches for the histogram.
plt.figure(figsize=(12, 6))

# Define colors for default ('Yes') and non-default ('No') cases.
default_colors = ['#012030', '#DAFDBA']

# Create separate histograms for default and non-default cases:
for credit_default, color in zip(['Yes', 'No'], default_colors):
    # Subset the data for the specific credit_default status.
    subset = data[data['Credit_default'] == credit_default]
    
    # Create a histogram using Seaborn.
    sns.histplot(data=subset, x='Age', element='step', common_norm=False, discrete=True, stat='count', color=color, label=f'Credit Default: {credit_default}')

# Set a title for the plot.
plt.title('Credit Default by Age Group')

# Label the x-axis and y-axis.
plt.xlabel('Age Group')
plt.ylabel('Count')

# Add a legend to distinguish between 'Yes' and 'No' credit defaults.
plt.legend()

# Display the plot.
plt.show()
````

## Encode Categorical Variables

To work effectively with machine learning algorithms, it's often necessary to convert categorical variables into numerical format. This process allows algorithms to process and analyze these variables accurately. Two common techniques for encoding categorical variables are one-hot encoding and label encoding.

### One-Hot Encoding (Recommended for Nominal Categorical Variables)

One-hot encoding is the preferred method for nominal categorical variables, where there is no inherent order or ranking among categories. This technique creates binary columns for each category, indicating the presence or absence of that category in each data point. Here's how it works:

1. **Identification of Categorical Variables**: Begin by identifying the categorical variables in your dataset. These are typically non-numeric columns that represent categories or groups.

2. **Application of One-Hot Encoding**: For each categorical variable, apply one-hot encoding. This process creates binary (0 or 1) columns for each category within that variable.

3. **Example**: If you have a categorical variable like "Color" with categories "Red," "Blue," and "Green," one-hot encoding would result in three binary columns: "Is_Red," "Is_Blue," and "Is_Green." For each data point, the corresponding column is set to 1 for the color it represents, and the others are set to 0.

One-hot encoding is especially useful when dealing with categorical variables that have no inherent order or ranking, as it prevents any bias introduced by numeric labels.

### Label Encoding

Label encoding is typically used for ordinal categorical variables, where there is a clear order or ranking among categories. This technique assigns numeric labels to the categories based on their order, allowing for their representation as ordinal values.

### Conclusion

The choice between one-hot encoding and label encoding depends on the nature of the categorical variable and the requirements of the analysis. One-hot encoding is recommended for nominal categorical variables, while label encoding is suitable for ordinal variables with a clear order.


````python
# One-Hot Encoding for Categorical Variables

# Specify the list of categorical columns to one-hot encode
categorical_columns = ['Gender', 'Living_Status', 'Car_two_wheeler', 'Residence', 'Loan_type', 'Bank_name']

# Apply one-hot encoding using pandas get_dummies method
data = pd.get_dummies(data, columns=categorical_columns)

# The 'data' DataFrame now contains one-hot encoded columns for the specified categorical variables.
````

## Feature Scaling: Min-Max Scaling

In machine learning, feature scaling is a crucial preprocessing step that ensures numerical features are on a similar scale. This is essential because many machine learning algorithms are sensitive to the magnitude of input features. One common technique for feature scaling is Min-Max scaling.

### How Min-Max Scaling Works

Min-Max scaling, also known as normalization, transforms numerical features so that they fall within a specific range, typically [0, 1]. This technique is especially useful when you want to maintain the original data distribution and ensure all features have equal influence. Here's how Min-Max scaling works:

1. **Identification of Numerical Features**: Begin by identifying the numerical features in your dataset that need to be scaled.

2. **Min-Max Scaling Formula**: To scale a feature, you use the following formula for each data point (x):

`x_scaled = (x - min(x)) / (max(x) - min(x))`

In this formula, 'x_scaled' is the scaled value, 'x' is the original value, 'min(x)' is the minimum value of the feature, and 'max(x)' is the maximum value of the feature.

3. **Applying Min-Max Scaling**: Apply this formula to each data point within the feature, transforming them to the [0, 1] range.

4. **Benefits**: Min-Max scaling ensures that all features contribute equally to the analysis, avoids potential issues with features having widely different scales, and allows machine learning algorithms to converge faster.

### Example

Here's an example of Min-Max scaling in Python using the `MinMaxScaler` from the `sklearn.preprocessing` library:


````python
# Import the necessary library for Min-Max scaling
from sklearn.preprocessing import MinMaxScaler

# Create a Min-Max scaler instance
scaler = MinMaxScaler()

# Apply Min-Max scaling to the 'Monthly_Income' column
data['Monthly_Income'] = scaler.fit_transform(data[['Monthly_Income']])

# Apply Min-Max scaling to the 'Loan_int_rate' column
data['Loan_int_rate'] = scaler.fit_transform(data[['Loan_int_rate']])

# You can use the same Min-Max scaling technique for other numerical columns as needed.
````

## Feature Scaling: Z-Score Standardization

In machine learning, feature scaling is a crucial preprocessing step to ensure that numerical features are on a similar scale. One common technique for feature scaling is Z-score standardization, which transforms data to have a mean of 0 and a standard deviation of 1. This technique is particularly useful for a variety of reasons.

### How Z-Score Standardization Works

Z-score standardization, also known as zero-mean normalization, involves rescaling features to have a mean (average) of 0 and a standard deviation (a measure of dispersion) of 1. Here's how Z-score standardization works:

1. **Identification of Numerical Features**: Start by identifying the numerical features in your dataset that require scaling.

2. **Z-Score Formula**: To standardize a feature, you apply the following formula to each data point (x):

`z = (x - μ) / σ`


In this formula, 'z' is the standardized value, 'x' is the original value, 'μ' is the mean of the feature, and 'σ' is the standard deviation of the feature.

3. **Applying Z-Score Standardization**: Apply this formula to each data point within the feature, transforming them to have a mean of 0 and a standard deviation of 1.

### Benefits of Z-Score Standardization

Z-score standardization offers several benefits, including:

- **Mean Centering**: By setting the mean to 0, it ensures that data is centered around the origin, making it easier to interpret.

- **Scaling for Outliers**: Z-score standardization is robust to outliers, as it relies on the mean and standard deviation, which are less affected by extreme values.

- **Standardized Units**: It transforms data into standardized units, allowing for better comparisons between features with different units or scales.

- **Improved Algorithm Performance**: Machine learning algorithms often perform better when features are standardized. Algorithms like k-means clustering and support vector machines rely on distances between data points, and standardization ensures that all features contribute equally to these calculations.

- **Normal Distribution**: Z-score standardization can make data more closely resemble a normal distribution, which can be beneficial for certain statistical tests.

### Example

Here's an example of Z-score standardization in Python using the `StandardScaler` from the `sklearn.preprocessing` library:


````python
# Import the necessary library for Z-score standardization
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
scaler = StandardScaler()

# Apply Z-score standardization to the 'Monthly_Income' column
data['Monthly_Income'] = scaler.fit_transform(data[['Monthly_Income']])

# Apply Z-score standardization to the 'Loan_int_rate' column
data['Loan_int_rate'] = scaler.fit_transform(data[['Loan_int_rate']])

# You can use the same Z-score standardization technique for other numerical columns as needed.
````

To assess the model's performance, we evaluate several key metrics:

- **Accuracy**: Measures the proportion of correctly classified instances.
- **Precision**: Indicates the ratio of true positive predictions to the total positive predictions.
- **Recall**: Represents the ratio of true positive predictions to the total actual positive instances.
- **F1 Score**: Combines precision and recall, providing a balanced measure.
- **ROC-AUC Score**: Measures the area under the Receiver Operating Characteristic (ROC) curve, assessing the model's ability to distinguish between classes.

#### Step 4: Model Evaluation for Logistic Regression

````python
# Step 4: Model Evaluation for Logistic Regression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Predict on the testing data
lr_predictions = lr_model.predict(X_test_encoded)

# Evaluate the LR model
lr_accuracy = accuracy_score(y_test, lr_predictions)
lr_precision = precision_score(y_test, lr_predictions)
lr_recall = recall_score(y_test, lr_predictions)
lr_f1_score = f1_score(y_test, lr_predictions)
lr_roc_auc = roc_auc_score(y_test, lr_predictions)

# Print the evaluation metrics for Logistic Regression
print("Logistic Regression Model Evaluation:")
print(f"Accuracy: {lr_accuracy:.2f}")
print(f"Precision: {lr_precision:.2f}")
print(f"Recall: {lr_recall:.2f}")
print(f"F1 Score: {lr_f1_score:.2f}")
print(f"ROC-AUC: {lr_roc_auc:.2f}")
````

## Classification Report: Understanding Model Performance

A classification report is a valuable tool for assessing the performance of a classification model. It provides a comprehensive summary of various performance metrics for each class within a classification problem. Here's a breakdown of the key components of the classification report and why it is essential:

### Precision
Precision is a metric that measures the accuracy of positive predictions made by the model. It's calculated as the ratio of true positives to the total number of positive predictions. High precision indicates that the model makes fewer false-positive errors, making it particularly valuable in situations where false positives are costly or undesirable.

### Recall (Sensitivity)
Recall, also known as sensitivity or true positive rate, assesses the model's ability to identify all relevant instances within a dataset. It's computed as the ratio of true positives to the total number of actual positives. High recall values signify that the model is successful at minimizing false negatives, which is crucial in cases where missing relevant instances is problematic.

### F1-Score
The F1-score is the harmonic mean of precision and recall. It offers a balanced measurement of precision and recall, making it particularly useful when the class distribution is imbalanced or when both precision and recall need to be considered equally.

### Support
Support represents the number of actual occurrences of each class in the dataset. It provides insights into the class distribution and the number of samples associated with each class.

### Why Do We Need a Classification Report?
A classification report is essential for the following reasons:

- **Comprehensive Evaluation:** It offers a more comprehensive view of the model's performance by providing multiple performance metrics for each class, enabling us to assess strengths and weaknesses in class-specific predictions.

- **Class-Dependent Assessment:** In problems with multiple classes, a classification report is invaluable for understanding how the model performs for each class. It identifies classes that may have lower accuracy or are more challenging to predict accurately.

- **Decision-Making:** The report aids in making informed decisions based on the specific requirements of the classification problem. It helps assess whether the model meets the desired precision, recall, or F1-score thresholds for each class.

In summary, the classification report is a critical tool for evaluating and fine-tuning classification models, particularly in scenarios involving multiple classes. It goes beyond a single accuracy score, providing a detailed assessment of the model's performance, allowing for more informed decisions and improvements.

## Learning Curve for Logistic Regression

In this section, we create a learning curve to visualize the performance of a logistic regression model as the training dataset size increases. Learning curves are valuable for understanding how model performance changes with the availability of more training data.

### Learning Curve Generation

- We utilize the `learning_curve` function from scikit-learn to create the learning curve. This function calculates training and cross-validation scores for varying training dataset sizes.

- The `train_sizes` variable represents the different sizes of the training dataset, while `train_scores` and `test_scores` store the corresponding training and cross-validation scores.

### Visualization

- The learning curve is visualized using a line plot. It displays the trend of training and cross-validation scores as the training dataset size grows.

- The red line and shaded region represent the training score (accuracy) and its standard deviation. The green line and shaded region represent the cross-validation score (accuracy) and its standard deviation.

### Interpretation

- A learning curve reveals essential insights about a model:
  - If there is a large gap between the training and cross-validation scores with a small dataset, it suggests overfitting.
  - If both scores are low, the model may be underfitting, and more data or model complexity is needed.
  - If both scores converge and are close to each other, it indicates a well-generalized model.

- As the training dataset size increases, the model's performance can be assessed in terms of bias and variance trade-off. Smaller datasets may lead to more variance (overfitting), while larger datasets tend to reduce variance but may suffer from bias (underfitting).

### Conclusion

Learning curves are useful tools for understanding the impact of dataset size on logistic regression model performance. Analyzing these curves can help in making informed decisions about data collection and model development.


````python
import numpy as np  # Add this line to import NumPy

from sklearn.model_selection import learning_curve

# Create a learning curve
train_sizes, train_scores, test_scores = learning_curve(lr_model, X_train_encoded, y_train, cv=5)

# Calculate the mean and standard deviation of training and testing scores
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plot the learning curve
plt.figure(figsize=(8, 5))
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color='r')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color='g')
plt.plot(train_sizes, train_mean, 'o-', color='r', label='Training Score')
plt.plot(train_sizes, test_mean, 'o-', color='g', label='Cross-Validation Score')
plt.title('Learning Curve')
plt.xlabel('Training Examples')
plt.ylabel('Score')
plt.legend(loc='best')
plt.show()
````

## Cross Validation Scoring

Cross-validation is a crucial technique in machine learning to assess a model's performance and ensure it generalizes well to unseen data. It helps prevent issues such as overfitting and provides a more reliable estimate of a model's effectiveness.

### Cross Validation Overview

- Cross-validation involves dividing the dataset into multiple subsets or "folds." The model is trained on a portion of the data and validated on the remaining portion.
- One common form of cross-validation is k-fold cross-validation, where the dataset is divided into k subsets. The model is trained k times, with each subset serving as the validation set exactly once.
- Cross-validation allows you to evaluate a model's performance on different portions of the data, providing a more robust estimate of how the model will perform on unseen data.

### Cross Validation Scoring

- In the given code, cross-validation is performed using the `cross_val_score` function from scikit-learn.
- The code assumes you have already trained a logistic regression model, `lr_model`, on your data.
- The `cross_val_score` function takes the following parameters:
  - `lr_model`: The trained model you want to evaluate.
  - `X` and `y`: The features and target variable, respectively.
  - `cv`: The number of folds, in this case, 5-fold cross-validation.
  - `scoring`: The scoring metric used to assess the model's performance, in this case, 'accuracy.'

### Output Interpretation

- The code prints the cross-validation scores for each fold. These scores represent the accuracy of the model on different validation sets.
- The 'Mean Accuracy' is computed by taking the average of the cross-validation scores. It provides an overall estimate of the model's performance.

### Benefits of Cross Validation

- Cross-validation is essential for estimating a model's generalization performance and detecting issues like overfitting.
- It helps in optimizing model parameters, as you can tune the model based on cross-validation results.
- Cross-validation ensures that your model is robust and reliable for making predictions on new, unseen data.

By using cross-validation, you gain more confidence in the model's performance, making it a valuable practice in machine learning model assessment and selection.

## K-Fold Cross Validation for Deep Neural Networks

K-fold cross-validation is a widely used technique in machine learning for assessing the performance of models, including Deep Neural Networks (DNNs). In this code snippet, K-fold cross-validation is applied to a trained DNN model.

### Cross Validation Overview

- Cross-validation involves dividing the dataset into multiple subsets or "folds." The model is trained and validated on different subsets in a repetitive manner.
- In this code, K-fold cross-validation is performed using the `KFold` class from scikit-learn. It divides the data into `n_splits` (in this case, 5) folds.
- The code iterates over each fold, training and evaluating the DNN model on different portions of the data.

### Code Explanation

- The code assumes that you have already trained a DNN model, `dnn_model`, and have your features `X` and target variable `y`.

1. **KFold Initialization**:
   - The number of folds for cross-validation (`n_splits`) is defined as 5.
   - A `KFold` object `kf` is created with this number of splits.

2. **Cross-Validation Scores List**:
   - A list `cross_val_scores_dnn` is initialized to store the accuracy scores obtained in each fold.

3. **Cross-Validation Loop**:
   - The code enters a loop that iterates through each fold:
     - The data is split into training and testing sets using the indices provided by `kf.split(X)`.
     - The DNN model is trained on the training data for a specified number of epochs and batch size.
     - The model is evaluated on the testing data, and the accuracy score is computed.
     - The accuracy score for the current fold is appended to the `cross_val_scores_dnn` list.

4. **Printing Results**:
   - After all folds are processed, the code prints the cross-validation scores for each fold.
   - The 'Mean Accuracy' is computed by taking the average of the cross-validation scores, providing an overall estimate of the model's performance.

### Benefits of K-Fold Cross Validation

- K-fold cross-validation is a robust technique to assess a model's generalization performance and identify issues such as overfitting.
- It helps in optimizing model hyperparameters, as you can tune the model based on cross-validation results.
- By performing cross-validation, you gain more confidence in the model's performance, making it a valuable practice in machine learning model evaluation.

This code snippet illustrates how to implement K-fold cross-validation for a DNN model, ensuring that the model's performance is rigorously evaluated and providing a reliable estimate of its effectiveness on unseen data.


````python
from sklearn.model_selection import KFold

# Assuming you have already trained your DNN model dnn_model

# Define the number of folds for cross-validation
n_splits = 5
kf = KFold(n_splits=n_splits)

# Initialize a list to store the cross-validation scores
cross_val_scores_dnn = []

# Perform cross-validation
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Train and evaluate the model on the current fold
    dnn_model.fit(X_train, y_train, epochs=10, batch_size=32)
    _, accuracy = dnn_model.evaluate(X_test, y_test)
    cross_val_scores_dnn.append(accuracy)

# Print the cross-validation scores
print("DNN Cross-Validation Scores:", cross_val_scores_dnn)
print("Mean Accuracy:", np.mean(cross_val_scores_dnn))
````

## Stratified K-Fold Cross Validation for Deep Neural Networks

In machine learning, cross-validation is a vital technique to assess the performance and reliability of models. Stratified K-Fold cross-validation is particularly useful when dealing with imbalanced datasets. In this code snippet, we apply Stratified K-Fold cross-validation to a trained Deep Neural Network (DNN) model and evaluate its performance.

### Cross Validation Overview

- Cross-validation helps in evaluating how well a model generalizes to unseen data by partitioning the dataset into subsets or "folds."
- Stratified K-Fold cross-validation ensures that each fold maintains the same class distribution as the whole dataset, which is crucial for imbalanced datasets.

### Code Explanation

- The code assumes that you have already defined and trained a DNN model (`dnn_model`) and have your features (`X`) and target variable (`y`).

1. **Model Definition and Compilation**:
   - The architecture of the DNN model is defined, including the number of layers, activation functions, and loss function.
   - The model is compiled with the 'adam' optimizer and 'binary_crossentropy' loss function for binary classification.

2. **Data Preparation**:
   - The target variable `y` is converted to a 1D NumPy array (`y_1d`) using `.to_numpy().ravel()`. This step is necessary for compatibility with some scikit-learn functions.

3. **Stratified K-Fold Cross-Validation**:
   - The `StratifiedKFold` method is initialized with 5 splits (`n_splits=5`), shuffling the data and using a specific random state.

4. **Metric Lists**:
   - Lists are initialized to store the precision, recall, F1 score, and ROC AUC score for each fold.

5. **Cross-Validation Loop**:
   - The code enters a loop that iterates through each fold.
   - The data is split into training and testing sets using the indices provided by `cv.split(X, y_1d)`.
   - The DNN model is trained on the training data for 10 epochs with a batch size of 32, suppressing output (`verbose=0`).
   - Predictions are made on the testing data, and metrics (precision, recall, F1 score, and ROC AUC) are calculated for each fold.

6. **Results Display**:
   - After all folds are processed, the code prints the average scores and standard deviation for each metric:
     - Precision
     - Recall
     - F1 Score
     - ROC AUC
   - The format includes the mean score and a measure of variation (+/- two times the standard deviation).

### Benefits of Stratified K-Fold Cross Validation

- Stratified K-Fold cross-validation helps in ensuring reliable model evaluation on imbalanced datasets by preserving class distribution.
- It provides an estimate of the model's performance that is less sensitive to the data split.

This code snippet demonstrates how to perform Stratified K-Fold cross-validation for a DNN model and assess its performance, especially when dealing with imbalanced datasets. It allows you to confidently evaluate and validate the model's ability to generalize to new, unseen data.


````python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
from sklearn.model_selection import StratifiedKFold
from tensorflow import keras

# Define the model architecture
dnn_model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(X.shape[1],)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

# Compile the model
dnn_model.compile(optimizer='adam', loss='binary_crossentropy')

# Assuming you have already trained your DNN model 'dnn_model'
# Convert y to a NumPy array
y_1d = y.to_numpy().ravel()

# Define the cross-validation method (Stratified K-Fold)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize lists to store evaluation metrics
precision_scores = []
recall_scores = []
f1_scores = []
roc_auc_scores = []

# Perform cross-validation
for train_idx, test_idx in cv.split(X, y_1d):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y_1d[train_idx], y_1d[test_idx]

    # Train and evaluate the DNN model on each fold
    dnn_model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)  # Add 'verbose' to suppress output
    y_pred = dnn_model.predict(X_test)

    precision_scores.append(precision_score(y_test, y_pred > 0.5))
    recall_scores.append(recall_score(y_test, y_pred > 0.5))
    f1_scores.append(f1_score(y_test, y_pred > 0.5))
    roc_auc_scores.append(roc_auc_score(y_test, y_pred))

# Print the average scores and standard deviation for each metric
print("DNN Model - Precision: {:.4f} (+/- {:.4f})".format(np.mean(precision_scores), np.std(precision_scores) * 2))
print("DNN Model - Recall: {:.4f} (+/- {:.4f})".format(np.mean(recall_scores), np.std(recall_scores) * 2))
print("DNN Model - F1 Score: {:.4f} (+/- {:.4f})".format(np.mean(f1_scores), np.std(f1_scores) * 2))
print("DNN Model - ROC AUC: {:.4f} (+/- {:.4f})".format(np.mean(roc_auc_scores), np.std(roc_auc_scores) * 2))
````
